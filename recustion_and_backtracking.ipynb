# Recursion

```python
def print_num(n):
    if n == 0:
        return
    print_num(n - 1)
    print(n)

print(print_num(5))
```

# Factorial

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n -1)

print(factorial(5))
```

```python
def test(n):
    if n == 0:
        return 
    print(n)
    test(n - 1)
    print(n)
test(3)
```
