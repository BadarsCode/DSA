# Linear Search

```python
import random 
arr = [random.randint(0, 10000) for _ in range(100_000)]
print(arr)

target = 7 

for i in range(len(arr)):
    if arr[i] == target:
        print(f"found at index {i}")
        break
```

# BINARY SEARCH

```python
arr = [2,3,5,7,9,13,19,29,30]
target =29

left = 0
right = len(arr) - 1 
count = 0
while left <= right:
    count += 1
    print(f"iteration: {count}")
    mid = (left + right) // 2

    if arr[left] == target:
        print(f"found at index {mid}")
        break 
    elif arr[mid] < target:
        left = mid + 1 
    else:
        right = mid - 1
```
