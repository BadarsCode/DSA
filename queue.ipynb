# Queue

```python
# Queue

class Queue:
    def __init__(self):
        self.q=[]
    def enqueue(self, data):
        self.q.append(data)
    def dequeue(self):
        if not self.q:
            return None 
        return self.q.pop(0)
    def front(self):
        if not self.q:
            return None
        return self.q[0]
    def isEmpty(self):
        return len(self.q) == 0
    
```

```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue())
```

```python
print(q.front())
```

```python
print(q.isEmpty())
```

```python
print(q.front())
```

# queue using two stack

```python
class Two_Stack_Queue:
    def __init__(self):
        self.outstack = []
        self.instack = []
    def push(self, data):
        self.instack.append(data)
    def pop(self):
        self.move()
        return self.outstack.pop()
    def peek(self):
        self.move()
        return self.outstack[-1]
    def move(self):
        if not self.outstack:
            while self.instack:
                self.outstack.append(self.instack.pop())
    def show_instack(self):
        instack = self.instack
        return instack
    def show_outstack(self):
        outstack = self.outstack
        return outstack


```

```python
q = Two_Stack_Queue()
q.push(1)
q.push(2)
q.push(3)
print(q.pop())
print(q.peek())
print(q.show_instack())
print(q.show_outstack())
```

# Double ended Queue

```python
from collections import deque

dq= deque()

dq.append(1)
dq.appendleft(2)
dq.append(3)
print(dq)
```

```python
dq.pop()
print(dq)
```

```python
dq.popleft()
print(dq)
```
