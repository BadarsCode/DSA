# Queue

```python
# Queue

class Queue:
    def __init__(self):
        self.q=[]
    def enqueue(self, data):
        self.q.append(data)
    def dequeue(self):
        if not self.q:
            return None 
        return self.q.pop(0)
    def front(self):
        if not self.q:
            return None
        return self.q[0]
    def isEmpty(self):
        return len(self.q) == 0
    
```

```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue())
```

```python
print(q.front())
```

```python
print(q.isEmpty())
```

```python
print(q.front())
```
