# Graph 

## depth first search

```python
def dfs(graph, node, visited=set()):
    if node in visited:
        return
    print(node , end=" ")
    visited.add(node)
    for neighbor in graph[node]:
        dfs(graph, neighbor, visited)
```

```python
print("Depth-First Search (DFS) Traversal:")
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': []
}

dfs(graph, 'A')
```
