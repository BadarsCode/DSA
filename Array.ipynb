```python
# access
arr = [12,23,24,65]
print(arr[1])
```

```python
# append
arr.append(45)
print(arr)
```

```python
# update
arr[2] = 100
print(arr)
```

```python
# find maximum number in array
max_number = max(arr)
print("Maximum number is:", max_number)
```

```python
# using loop
max_number = arr[0]
for num in arr:
    if num > max_number:
        max_number = num
print(f'Maximum number using loop is :{max_number}')
```

```python
# Sum of all numbers in array
total = sum(arr)
print(f"Sum of all numbers in array is: {total}")
```

```python
# using loop 
total = 0 
for num in arr:
    total += num
print(f"sum of all the numbers using loop is: {total}")
```

```python
def reverse_array(array):
    left = 0 
    right = len(array) -1

    for i in range(len(array)//2):
        array[left], array[right] = array[right], array[left]
        left += 1
        right -= 1
    return array
print(f"Original array is :{arr}")
reversed_arr = reverse_array(arr)
print(f"Reversed array is :{reversed_arr}")
```

# sliding window

```python
# sliding window maximum
def maximum_sum_subarray(array, k):
    window_sum=sum(array[0:k])
    max_sum = window_sum

    start = 0
    for end in range(k, len(array)):
        # slide the window 
        window_sum += array[end]
        window_sum -= array[start]
        start += 1
        max_sum = max(max_sum, window_sum)
        

    return max_sum

arr = [2,5,3,7,6,4,8,1,9]
maximum_sum_subarray(arr, 3)
```
