# dynamic programming

## Fibonacci

```python
import time 
start_time = time.time()
def fib(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n -1 ) + fib(n-2)
end_time = time.time()
print(fib(30))
print(f"Total time taken: {end_time - start_time} seconds")
```

```python
print(fib(10))
print(fib((20)))
print(fib(5))
```

## fibonacci series good (memorization)

```python
start_time = time.time()
def fib_memo(n, memo={}):
    if n <= 1:
        return n 
    if n in memo:
        return memo[n]
    memo[n] = fib_memo(n -1 , memo) + fib_memo(n -2, memo)
    return memo[n]
end_time = time.time()
print(fib_memo(30))
print(f"total time taken:{end_time - start_time}  seconds")
```

```python
print(fib_memo(10))
```
