```python
from typing import List

class Solution:
    def largestSubmatrix(self, matrix: List[List[int]]) -> int:
        
        m, n = len(matrix), len(matrix[0])
        
        # Step 1: Build heights
        for i in range(1, m):
            for j in range(n):
                if matrix[i][j] == 1:
                    matrix[i][j] += matrix[i-1][j]
        
        max_area = 0
        
        # Step 2 & 3
        for row in matrix:
            row.sort(reverse=True)
            
            for j in range(n):
                height = row[j]
                width = j + 1
                max_area = max(max_area, height * width)
        
        return max_area
```

```python
sol = Solution()
    
matrix = [[0,0,1],[1,1,1],[1,0,1]]
    
result = sol.largestSubmatrix(matrix)
    
print("Output:", result)
```

# add two numbers

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def addTwoNumbers(self, l1, l2):
        
        dummy = ListNode(0)
        current = dummy
        
        carry = 0
        
        while l1 or l2 or carry:
            
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0
            
            total = val1 + val2 + carry
            
            carry = total // 10
            digit = total % 10
            
            current.next = ListNode(digit)
            current = current.next
            
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next
        
        return dummy.next
```

```python
def create_linked_list(arr):
    dummy = ListNode(0)
    current = dummy
    
    for num in arr:
        current.next = ListNode(num)
        current = current.next
    
    return dummy.next

def print_linked_list(head):
    result = []
    
    while head:
        result.append(head.val)
        head = head.next
    
    print(result)
sol = Solution()
    
l1 = create_linked_list([2,4,3])
l2 = create_linked_list([5,6,4])
    
result = sol.addTwoNumbers(l1, l2)
    
print_linked_list(result)
```

# median of two sorted array

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        
        # Ensure nums1 is smaller
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1
        
        x, y = len(nums1), len(nums2)
        left, right = 0, x
        
        while left <= right:
            
            partitionX = (left + right) // 2
            partitionY = (x + y + 1) // 2 - partitionX
            
            maxLeftX = float('-inf') if partitionX == 0 else nums1[partitionX - 1]
            minRightX = float('inf') if partitionX == x else nums1[partitionX]
            
            maxLeftY = float('-inf') if partitionY == 0 else nums2[partitionY - 1]
            minRightY = float('inf') if partitionY == y else nums2[partitionY]
            
            # Correct partition
            if maxLeftX <= minRightY and maxLeftY <= minRightX:
                
                if (x + y) % 2 == 0:
                    return (max(maxLeftX, maxLeftY) + min(minRightX, minRightY)) / 2
                else:
                    return max(maxLeftX, maxLeftY)
            
            elif maxLeftX > minRightY:
                right = partitionX - 1
            else:
                left = partitionX + 1
```

```python
sol = Solution()
    
print(sol.findMedianSortedArrays([1,3], [2]))      
print(sol.findMedianSortedArrays([1,2], [3,4]))     
```

# Longest substring palindrome

```python
def longestPalindrome(s: str) -> str:
    
    def expand(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        
        return s[left+1:right]
    
    
    result = ""
    
    for i in range(len(s)):
        
        # Odd length
        p1 = expand(i, i)
        
        # Even length
        p2 = expand(i, i+1)
        
        # Take longest
        if len(p1) > len(result):
            result = p1
        
        if len(p2) > len(result):
            result = p2
    
    return result
```

```python
print(longestPalindrome("babad"))
print(longestPalindrome("cbbd"))
```

# Reverse Integer

```python
def reverse(x: int) -> int:
    
    sign = -1 if x < 0 else 1
    x = abs(x)
    
    rev = 0
    
    while x != 0:
        digit = x % 10
        x = x // 10
        
        rev = rev * 10 + digit
    
    rev *= sign
    
    # 32-bit check
    if rev < -2**31 or rev > 2**31 - 1:
        return 0
    
    return rev
```

```python
print(reverse(123))    
print(reverse(-123))   
print(reverse(120))
print(reverse(1534236469)) 
```

# String to Integer

```python
def myAtoi(s: str) -> int:
    
    s = s.lstrip()   # remove leading spaces
    
    if not s:
        return 0
    
    sign = 1
    i = 0
    
    # Check sign
    if s[i] == '-':
        sign = -1
        i += 1
    elif s[i] == '+':
        i += 1
    
    num = 0
    
    # Process digits
    while i < len(s) and s[i].isdigit():
        
        digit = int(s[i])
        
        # Overflow check
        if num > (2**31 - 1 - digit) // 10:
            return 2**31 - 1 if sign == 1 else -2**31
        
        num = num * 10 + digit
        i += 1
    
    return sign * num
```

```python
print(myAtoi("42"))
print(myAtoi("   -42"))
print(myAtoi("4193 with words"))
print(myAtoi("words 123"))
print(myAtoi("-91283472332"))
```

# Valid Parentheses

```python
def isValid(s: str) -> bool:
    
    stack = []
    
    mapping = {
        ')': '(',
        ']': '[',
        '}': '{'
    }
    
    for char in s:
        
        if char in mapping:  # closing bracket
            
            if not stack or stack[-1] != mapping[char]:
                return False
            
            stack.pop()
        
        else:
            stack.append(char)
    
    return len(stack) == 0
```

```python
print(isValid("()"))
print(isValid("()[]{}"))
print(isValid("(]"))
print(isValid("([)]"))
print(isValid("{[]}"))
```

# sliding window

```python
def max_sum_subarray(arr, k):
    window_sum = sum(arr[: k])
    max_sum = window_sum 

    for i in range(k, len(arr)):
        window_sum += arr[i]
        window_sum -= arr[i - k] 
        max_sum = max(max_sum, window_sum)
    return max_sum 

```

```python
arr = [1, 4, 2, 10, 23, 3, 1, 0, 20]
k = 3

print(max_sum_subarray(arr, k))
```

# Longest Substring without repeating characters

```python
def longest_substring(s):
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1 
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
    return max_length
```

```python
st = "abcdeabdesce"
print(longest_substring(st))
```

```python
def two_sum(arr, target):
    left = 0 
    right = len(arr) - 1
    
    while left < right:
        if arr[left] + arr[right] == target:
            return [left, right]
        elif arr[left] + arr[right] < target:
            left += 1
        else:
            right -= 1
    return "not found"
```

```python
def remove_duplicate(arr):
    slow = 0
    for fast in range(1, len(arr)):
        if arr[fast] != arr[slow]:
            slow += 1
            arr[slow] = arr[fast]
    return slow + 1
print(remove_duplicate([1,1,2,2,3]))

```

# Zigzag Conversion

```python
def convert(s: str, numRows:int) -> str:
    if numRows == 1 or numRows >= len(s):
        return s
    
    rows = ["\n"] * numRows

    currentRow = 0 
    going_down = False 

    for char in s:
        rows[currentRow] += char
        
        if currentRow == 0 or currentRow == numRows - 1:
            going_down = not going_down 
        
        currentRow += 1 if going_down else -1 
    return "".join(rows)
```

```python
print(convert("PAYPALISHIRING", 3))
print(convert("PAYPALISHIRING", 4))
print(convert("A", 1))
```

# longest common prefix

```python
def longestCommonPrefix(strs):
    
    if not strs:
        return ""
    
    prefix = strs[0]
    
    for word in strs[1:]:
        
        while not word.startswith(prefix):
            prefix = prefix[:-1]
            
            if prefix == "":
                return ""
    
    return prefix
```

```python
print(longestCommonPrefix(["flower","flow","flight"])) 
print(longestCommonPrefix(["dog","racecar","car"]))
print(longestCommonPrefix(["product", 'produce', "programming"]))     
```

# group Anagrams

```python
from collections import defaultdict

def groupAnagrams(strs):
    
    groups = defaultdict(list)
    
    for word in strs:
        key = "".join(sorted(word))
        groups[key].append(word)
    
    return list(groups.values())

print(groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
```

# 3 Sum

```python
def threeSum(nums):
    
    nums.sort()
    result = []
    
    for i in range(len(nums)):
        
        # Skip duplicates
        if i > 0 and nums[i] == nums[i-1]:
            continue
        
        left = i + 1
        right = len(nums) - 1
        
        while left < right:
            
            total = nums[i] + nums[left] + nums[right]
            
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                
                # Skip duplicates
                while left < right and nums[left] == nums[left+1]:
                    left += 1
                while left < right and nums[right] == nums[right-1]:
                    right -= 1
                
                left += 1
                right -= 1
            
            elif total < 0:
                left += 1
            else:
                right -= 1
    
    return result
```

```python
print(threeSum([-1,0,1,2,-1,-4]))
```

# Regular Expression Matching

```python
def isMatch(s: str, p: str) -> bool:
    
    m, n = len(s), len(p)
    
    dp = [[False] * (n + 1) for _ in range(m + 1)]
    
    dp[0][0] = True
    
    # Handle patterns like a*, a*b*, etc.
    for j in range(2, n + 1):
        if p[j-1] == '*':
            dp[0][j] = dp[0][j-2]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            
            if p[j-1] == '.' or s[i-1] == p[j-1]:
                dp[i][j] = dp[i-1][j-1]
            
            elif p[j-1] == '*':
                
                # zero occurrence
                dp[i][j] = dp[i][j-2]
                
                # multiple occurrence
                if p[j-2] == '.' or p[j-2] == s[i-1]:
                    dp[i][j] = dp[i][j] or dp[i-1][j]
    
    return dp[m][n]
```

```python
print(isMatch("aa", "a"))
print(isMatch("aa", "a*"))
print(isMatch("ab", ".*"))
print(isMatch("aab", "c*a*b"))
```

# Wildcard matching

```python
def isMatch(s: str, p: str) -> bool:
    
    m, n = len(s), len(p)
    
    dp = [[False] * (n + 1) for _ in range(m + 1)]
    
    dp[0][0] = True
    
    # Handle leading '*'
    for j in range(1, n + 1):
        if p[j-1] == '*':
            dp[0][j] = dp[0][j-1]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            
            if p[j-1] == '?' or p[j-1] == s[i-1]:
                dp[i][j] = dp[i-1][j-1]
            
            elif p[j-1] == '*':
                dp[i][j] = dp[i][j-1] or dp[i-1][j]
    
    return dp[m][n]
```

```python
print(isMatch("abc", "a?c"))
print(isMatch("abc", "a*"))
print(isMatch("abc", "a*d"))
print(isMatch("adceb", "*a*b"))
```

# Edit Distance

```python
def minDistance(word1: str, word2: str) -> int:
    
    m, n = len(word1), len(word2)
    
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Base cases
    for i in range(m + 1):
        dp[i][0] = i   # delete all
    
    for j in range(n + 1):
        dp[0][j] = j   # insert all
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            
            if word1[i-1] == word2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            
            else:
                dp[i][j] = 1 + min(
                    dp[i][j-1],   # insert
                    dp[i-1][j],   # delete
                    dp[i-1][j-1]  # replace
                )
    
    return dp[m][n]

```

```python
print(minDistance("horse", "ros"))
print(minDistance("intention", "execution"))
```

# Memoization (top-down dp)

```python
def fib(n, dp={}):
    if n in dp:
        return dp[n]
    if n <= 1:
        return n 
    dp[n] = fib(n-1, dp) + fib(n-2, dp)
    print(f"dp[{n}] = {dp[n]}")
    display(dp)
    return dp[n]
```

# Tabulations (Bottom-up dp)

```python
def fib(n):
    if n <= 1:
        return n
    print(f"Calculating fib({n})")
    dp = [0]*(n+1)
    dp[0], dp[1] = 0, 1
    
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
        print(f"dp[{i}] = {dp[i]}")
    print(f"dp: {dp}")
    
    return dp[n]
```

```python
fib(6)
```

# space optimization

```python
def fib(n):
    if n <= 1:
        return n
    
    prev2 = 0
    prev1 = 1
    
    for i in range(2, n+1):
        curr = prev1 + prev2
        prev2 = prev1
        prev1 = curr
        print(f"fib({i}) = {curr}")
    
    return prev1
```

```python
print(fib(6))
```

# Climbing Stairs

```python
def climb(n, dp={}):
    if n <= 1:
        return 1
    
    if n in dp:
        return dp[n]
    
    dp[n] = climb(n-1, dp) + climb(n-2, dp)
    print(f"dp[{n}] = {dp[n]}")
    return dp[n]
```

```python
print(climb(5))
```

# min cost climbing stairs

```python
def minCost(cost):
    n = len(cost)
    dp = [-1]*n

    def f(i):
        if i <= 1:
            return cost[i]
        
        if dp[i] != -1:
            return dp[i]
        
        dp[i] = min(f(i-1), f(i-2)) + cost[i]
        return dp[i]

    return min(f(n-1), f(n-2))
```

```python
print(minCost([10, 15, 20]))
```

```python
def minCost2(cost):
    n = len(cost)
    dp = [0]*n

    dp[0] = cost[0]
    dp[1] = cost[1]
    print(f"dp[0] = {dp[0]}, dp[1] = {dp[1]}")
    for i in range(2, n):
        dp[i] = min(dp[i-1], dp[i-2]) + cost[i]
        print(f"dp[{i}] = {dp[i]}")

    return min(dp[n-1], dp[n-2])
```

```python
# space optimized version
def minCost3(cost):
    prev2 = cost[0]
    prev1 = cost[1]
    print(f"prev2 = {prev2}, prev1 = {prev1}")
    for i in range(2, len(cost)):
        curr = min(prev1, prev2) + cost[i]
        prev2 = prev1
        prev1 = curr
        print(f"curr = {curr}, prev2 = {prev2}, prev1 = {prev1}")

    return min(prev1, prev2)
```

# House Robber

```python
def rob(nums):
    n = len(nums)
    dp = [-1]*n

    def f(i):
        if i == 0:
            return nums[0]
        if i < 0:
            return 0
        
        if dp[i] != -1:
            return dp[i]
        
        pick = nums[i] + f(i-2)
        not_pick = f(i-1)
        
        dp[i] = max(pick, not_pick)
        return dp[i]

    return f(n-1)
```

```python
print(rob([1,2,3,1]))
```

```python
def rob2(nums):
    n = len(nums)
    dp = [0]*n

    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1])
    print(f"dp[0] = {dp[0]}, dp[1] = {dp[1]}")

    for i in range(2, n):
        pick = nums[i] + dp[i-2]
        not_pick = dp[i-1]
        dp[i] = max(pick, not_pick)
        print(f"dp[{i}] = {dp[i]}")

    return dp[n-1]
```

# Maximum Subarray (Kadane’s Algorithm)

```python
def maxSubArray(nums):
    curr_sum = nums[0]
    max_sum = nums[0]

    for i in range(1, len(nums)):
        curr_sum = max(nums[i], curr_sum + nums[i])
        max_sum = max(max_sum, curr_sum)

    return max_sum
```

```python
print(maxSubArray([-2,1,-3,4,-1,2,1,-5,4]))
```
